# ------- Binomial Distributions -------

<b>Binomial distribution</b> is a statistical probability distribution that states the likelihood that a value will take one of two independent values under a given set of parameters or assumptions.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

<b>#Question: </b>A bank issue credit to customer under the scheme of Master Card. Based on the past data, the bank has found out that 60% of all accounts pay on the bill, if a sample of 7 accounts is selected at random form the current database cunstruct the binomial probability distrubition of accounts paying on time.

```python
n = 7 #binomial Distribution
```

```python
p = .6 #Parameters
```

```python
k = np.arange(0,8) #Sample Space
k
```

```python
#Binomial Probability Mass Function - binom.pmf
binomial = stats.binom.pmf(k,n,p)
binomial
```

```python
plt.plot(k,binomial, 'o-')
plt.title('Binomial: n=%i, p=%.2f'% (n,p), fontsize=15)
plt.xlabel("Number of Success")
plt.ylabel("Probability of Success")
plt.show()
```

```python
n = len(binomial)
```

```python
n
```

```python
sum(binomial)
```

# ------- Poisson Distributions -------


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

<b>#Question: </b>If on an avarage 6 customers arrive every two minutes at a bank during the busy hours of working.
<b>a)</b> what is the probability that exactly four customer arrive in a given minute?
<b>b)</b> What is the probability that more than three customers will arrive in a given minute?

```python
rate = 3
n = np.arange(0,20)
poisson = stats.poisson.pmf(n,rate)
poisson
```

```python
poisson[4]
```

```python
1 - (poisson[0]+poisson[1]+poisson[2]+poisson[3])
```

```python
plt.plot(n,poisson,'o-')
plt.title('Poisson: $\Lambda$ = %i ' % rate)
plt.xlabel('Number of Customers arriving in a minute')
plt.ylabel('Probability of number of customers arriving in minute')
plt.show()
```

# ------- Normal Distributions -------


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

The mean weight of a morning breakfast cereal pack is 0.295 kg. The random variable of the pack follows a normal distribution.

a) What is the probability that the pack weight less than 0.280 kg?
b) What is the probability that the pack weight more than 0.350 kg?
c) What is the probability that the pack weight between 0.260 to 0.340kg?

```python
#Calculate the Z value
z = (0.280 - 0.295)/0.025
z
```

```python
#use the calculated z value to calculate the probability or the area under the left of this z value
stats.norm.cdf(z) #Normal Cumulative Distrubution Function - norm.cdf
```

```python
#use the original values to calculate the probability
stats.norm.cdf(0.280, loc=0.295, scale=0.025)
```

```python
#The probability (area) to the right is calculated as (1 - probability to the left)

1-stats.norm.cdf(0.350,loc=0.295,scale=0.025)
```

```python
#The probability between points can be calculated by taking differences
stats.norm.cdf(0.340,loc=0.295,scale=0.025)-stats.norm.cdf(0.260,loc=0.295,scale=0.025)
```

```python
#the probability of being between 1,2 and 3 standard diviation of the mean
stats.norm.cdf(1)-stats.norm.cdf(-1)
```

```python
stats.norm.cdf(2)-stats.norm.cdf(-2)
```

```python
stats.norm.cdf(3)-stats.norm.cdf(-3)
```
